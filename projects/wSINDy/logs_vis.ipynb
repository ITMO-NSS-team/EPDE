```python
from epde.interface.logger import LogLoader
```

```python
# Van-der-Pol oscillator
logloader = LogLoader(filename = 'C:/Users/Mike/Documents/Work/EPDE/logs/Van_der_Pol_noised.json')
terms = [frozenset(('u{power: 2.0}', 'du/dx1{power: 1.0}')), frozenset(('du/dx1{power: 1.0}',)), frozenset(('u{power: 1.0}',)), frozenset(('d^2u/dx1^2{power: 1.0}',))]
weighted_size = lambda x: np.size(x)/5. 
logframe = logloader.to_pandas(terms, metrics = [np.mean, np.var, weighted_size])
```

```python
# Van-der-Pol oscillator ann_based
logloader_ann = LogLoader(filename = 'C:/Users/Mike/Documents/Work/EPDE/logs/Van_der_Pol_ANN_noised.json')
weighted_size = lambda x: np.size(x)/5. 
print(weighted_size(np.array([1, 2, 3])))

logframe_ann = logloader_ann.to_pandas(terms, metrics = [np.mean, np.var, weighted_size])
logframe[0].loc['epde_0.005'] = logframe_ann[0].loc['epde_0.005']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
heatmap_kwargs = {'cmap': sns.color_palette("coolwarm_r", as_cmap=True), 'linewidths': 1}
sns.heatmap(data = logframe[0][[colname for colname in logframe[0].columns if 'disc_num' in colname]].fillna(0),
            annot = False, 
            **heatmap_kwargs)
sns.heatmap(data = logframe[0][[colname for colname in logframe[0].columns if 'disc_num' in colname]].fillna(0),
            annot = logframe[0][[colname for colname in logframe[0].columns if 'mean' in colname]],
            annot_kws={'va':'bottom'}, fmt='.3f', cbar = False,
            **heatmap_kwargs)
sns.heatmap(data = logframe[0][[colname for colname in logframe[0].columns if 'disc_num' in colname]].fillna(0),
            annot = logframe[0][[colname for colname in logframe[0].columns if 'var' in colname]],
            annot_kws={'va':'top'}, fmt='.5f', cbar = False,
            **heatmap_kwargs)        
plt.savefig('van-der-pol.png', bbox_inches='tight', dpi = 300)    
```

```python
# Lotka-Volterra equations
logloader = LogLoader(filename = 'C:/Users/Mike/Documents/Work/EPDE/logs/lotka_volterra_new_EPDE.json')
u_terms = [frozenset(('u{power: 1.0}',)), frozenset(('du/dx1{power: 1.0}',)), frozenset(('u{power: 1.0}', 'v{power: 1.0}'))]
v_terms = [frozenset(('v{power: 1.0}',)), frozenset(('dv/dx1{power: 1.0}',)), frozenset(('u{power: 1.0}', 'v{power: 1.0}'))]

weighted_size = lambda x: np.size(x)/10. 
logframe_u = logloader.to_pandas(u_terms, metrics = [np.mean, np.var, weighted_size], variable = 'u', variables = ['u', 'v'])
logframe_v = logloader.to_pandas(v_terms, metrics = [np.mean, np.var, weighted_size], variable = 'v', variables = ['u', 'v'])
```

```python
heatmap_kwargs = {'cmap': sns.color_palette("coolwarm_r", as_cmap=True), 'linewidths': 1}
sns.heatmap(data = logframe_u[0][[colname for colname in logframe_u[0].columns if 'disc_num' in colname]].fillna(0),
            annot = False, 
            **heatmap_kwargs)
sns.heatmap(data = logframe_u[0][[colname for colname in logframe_u[0].columns if 'disc_num' in colname]].fillna(0),
            annot = logframe_u[0][[colname for colname in logframe_u[0].columns if 'mean' in colname]],
            annot_kws={'va':'bottom'}, fmt='.3f', cbar = False,
            **heatmap_kwargs)
sns.heatmap(data = logframe_u[0][[colname for colname in logframe_u[0].columns if 'disc_num' in colname]].fillna(0),
            annot = logframe_u[0][[colname for colname in logframe_u[0].columns if 'var' in colname]],
            annot_kws={'va':'top'}, fmt='.5f', cbar = False,
            **heatmap_kwargs)        
plt.savefig('lotka_volterra_u.png', bbox_inches='tight', dpi = 300)    
```

```python
heatmap_kwargs = {'cmap': sns.color_palette("coolwarm_r", as_cmap=True), 'linewidths': 1}
sns.heatmap(data = logframe_v[0][[colname for colname in logframe_v[0].columns if 'disc_num' in colname]].fillna(0),
            annot = False, 
            **heatmap_kwargs)
sns.heatmap(data = logframe_v[0][[colname for colname in logframe_v[0].columns if 'disc_num' in colname]].fillna(0),
            annot = logframe_v[0][[colname for colname in logframe_v[0].columns if 'mean' in colname]],
            annot_kws={'va':'bottom'}, fmt='.3f', cbar = False,
            **heatmap_kwargs)
sns.heatmap(data = logframe_v[0][[colname for colname in logframe_v[0].columns if 'disc_num' in colname]].fillna(0),
            annot = logframe_v[0][[colname for colname in logframe_v[0].columns if 'var' in colname]],
            annot_kws={'va':'top'}, fmt='.5f', cbar = False,
            **heatmap_kwargs)        
plt.savefig('lotka_volterra_v.png', bbox_inches='tight', dpi = 300)    
```
