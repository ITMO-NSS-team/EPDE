```python
f = lambdify(x, (0.072357460529242*sin(1.72753437013566*x) + 1.0*cos(1.72753437013566*x))*exp(-0.125*x))
df = ((0.072357460529242*sin(1.72753437013566*x) + 1.0*cos(1.72753437013566*x))*exp(-0.125*x)).diff(x)
df2 = df.diff(x)
df = lambdify(x, df)
df2 = lambdify(x, df2)
f = np.vectorize(f)
df = np.vectorize(df)
df2 = np.vectorize(df2)
from epde.preprocessing.deriv_calculators import SpectralDeriv, PolynomialDeriv
sd = SpectralDeriv()
x = np.linspace(0, 10, 1000)
v = f(x)#+np.random.normal(0, 0.01, x.size)
dfunc = df(x)
dfunc2 = df2(x)
err_1=[]
err_2=[]
for i in range(1,1000):
    s= sd.spectral_derivative_high_ord(v,[x], max_order = 2, n = i)
    err1 = np.linalg.norm(dfunc[100:900]-s[0][1][100:900])
    err2 = np.linalg.norm(dfunc2[100:900]-s[1][1][100:900])
    err_1.append(err1)
    err_2.append(err2)
err_1 = np.array(err_1)
err_2 = np.array(err_2)
err_1
```

```python
from epde.preprocessing.deriv_calculators import SpectralDeriv, PolynomialDeriv
sd = SpectralDeriv()
t = np.linspace(0, 1, 101)
x = np.linspace(0, 1, 101)
v = np.loadtxt('E:\pde_mine/wave_sln_100.csv', delimiter = ',')
dx = np.load('E:\pde_mine/d_dxwaxe.npy')
dt = np.load('E:\pde_mine/d_dtwave.npy') 
d2x = np.load('E:\pde_mine/d2_dx2wave.npy')
d2t = np.load('E:\pde_mine/d2_dt2wave.npy')
err_1x = []
err_1t = []
for i in range(10,100):
    s_x = sd.spectral_derivative_high_ord(v, [t] , max_order = 2, n = i)[1][1]
    s_t = sd.spectral_derivative_high_ord(v.T, [t] , max_order = 2, n = i)[1][1]
    err1t = np.linalg.norm((s_t - d2t)[10:90][:,10:90].flatten())
    err1x = np.linalg.norm((s_x - d2x)[10:90][:,10:90].flatten())
    err_1t.append(err1t)
    err_1x.append(err1x)
err_1t
```
